```python
raw = open(os.getcwd()+r'/06_01.txt', encoding = 'utf8').read()# os.getcwd() computer 내 풀 디렉토리 받아오기, 
#encoding 지정 안해주면 파일 깨짐
raw
```

```python
print(type(raw))
print(len(raw))
print(raw[:75])
```

```python
tokens = nltk.word_tokenize(raw)
print(type(tokens))
print(len(tokens))#단어의 숫자
print(tokens[:10])#words들
```

```python
text = nltk.Text(tokens)
print(type(text))
print(text[:10])
text.collocations()#가능성 높은 단어의 조합을 보여줌
```

```python
text.concordance('great', 79, 25) #라인의 알파벳이 79개, 25개의 라인
```

```python
raw = nltk.corpus.gutenberg.raw('melville-moby_dick.txt')
fdist = nltk.FreqDist(raw)#raw string에서 가장 많이 나온것을 알려줌
print(fdist.most_common(5))#상위 5개

#[ch.lower() for ch in raw if ch.isalpha()] ; 속에 들어있는 것이 알파벳일때만 가져와라

fdist = nltk.FreqDist(ch.lower() for ch in raw if ch.isalpha())
print(fdist.most_common(5))
[char for (char, count) in fdist.most_common(5)]#총 5번 돔
```
