```python
raw = nltk.corpus.gutenberg.raw('melville-moby_dick.txt')
fdist = nltk.FreqDist(raw)#raw stringÏóêÏÑú Í∞ÄÏû• ÎßéÏù¥ ÎÇòÏò®Í≤ÉÏùÑ ÏïåÎ†§Ï§å
print(fdist.most_common(5))#ÏÉÅÏúÑ 5Í∞ú

#[ch.lower() for ch in raw if ch.isalpha()] ; ÏÜçÏóê Îì§Ïñ¥ÏûàÎäî Í≤ÉÏù¥ ÏïåÌååÎ≤≥ÏùºÎïåÎßå Í∞ÄÏ†∏ÏôÄÎùº

fdist = nltk.FreqDist(ch.lower() for ch in raw if ch.isalpha())
print(fdist.most_common(5))
[char for (char, count) in fdist.most_common(5)]#Ï¥ù 5Î≤à Îèî
```
